```python
name_mapping = [
    "box",
    "circularTorus",
    "cone",
    "coneOffset",
    "cylinder",
    "cylinderSlope",
    "dish",
    "mesh",
    "pyramid",
    "rectangularTorus",
    "sphere"
]
```

```python
def show_objects(num:int):
    base_path = '/home/workspace/geometry-classifier/data/'

    csv_test = base_path + f'photos_v3/photos_{num}.csv'

    df = pd.read_csv(csv_test)
    
    for i in range(1, 9):
        image = df.iloc[i+100, 2:].to_numpy()
        image = np.array(image, dtype=float).reshape(224,224)
        plt.figure(figsize=(20, 10))
        plt.imshow(image, cmap='gray')
        plt.title(name_mapping[df.iloc[i+100,1]])
        plt.axis('off')
        plt.show()


arr = [864, 704, 653, 612, 500, 20, 18, 9, 6 ]
for i in arr:
    show_objects(i)
```
